```python
import pandas as pd

# Read the dataset
df = pd.read_csv("rainfall.csv")  

# Convert "datetime" column to datetime format (DD-MM-YYYY)
df["datetime"] = pd.to_datetime(df["datetime"], format="%d-%m-%Y", errors="coerce")

# Set datetime as index
df.set_index("datetime", inplace=True)

# Extract temperature data
temperature_data = df["temp"]

# Train-Test Split
train = temperature_data.loc["2016-01-01":"2019-12-31"]
test = temperature_data.loc["2020-01-01":"2020-11-15"]

# Print the range to verify
print("Train Data Range:", train.index.min(), "to", train.index.max())
print("Test Data Range:", test.index.min(), "to", test.index.max())
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform Dickey-Fuller Test on train data
result = adfuller(train)
print(f"ADF Statistic: {result[0]}")
print(f"p-value: {result[1]}")

# If p-value < 0.05, the data is stationary, otherwise, apply differencing
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA model (Example: ARIMA(1,1,1))
model = ARIMA(train, order=(1, 1, 1))  # Adjust order based on your data
fitted_model = model.fit()

# Print model summary
print(fitted_model.summary())

# Plot diagnostics
fitted_model.plot_diagnostics(figsize=(10, 6))
plt.show()
```

```python
# Forecast using the fitted model
forecast = fitted_model.forecast(steps=len(test))

# Plot the predictions vs actual test data
plt.figure(figsize=(10, 6))
plt.plot(test.index, test, label="Actual Data", color='blue')
plt.plot(test.index, forecast, label="Predicted Data", color='red')
plt.title("Temperature Forecast")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Calculate errors
rmse = np.sqrt(mean_squared_error(test, forecast))
mae = mean_absolute_error(test, forecast)

print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```

```python
import pmdarima as pm
from sklearn.metrics import mean_squared_error
import numpy as np

# Fit the model automatically with auto_arima
model = pm.auto_arima(train, 
                      seasonal=False,   # Set to True for seasonal ARIMA (SARIMA)
                      m=1,              # Frequency of the series, set to 1 for non-seasonal
                      stepwise=True,    # Enables stepwise search for better performance
                      trace=True)       # Prints progress of the search

# Print summary of the model
print(model.summary())

# Forecasting
forecast = model.predict(n_periods=len(test))

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(test, forecast))
print(f"RMSE: {rmse}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

residuals = fitted_model.resid
plt.figure(figsize=(10, 5))
sns.histplot(residuals, kde=True)
plt.title("Residual Distribution")
plt.show()
```

```python
# Generate forecast for the next n periods
forecast_periods = len(test)  # or specify any number of periods for future forecasting
forecast = fitted_model.get_forecast(steps=forecast_periods)
forecast_mean = forecast.predicted_mean  # predicted values
forecast_conf_int = forecast.conf_int()  # confidence intervals

# Plot the forecasted values along with the actual data
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label="Training Data", color='blue')
plt.plot(test.index, test, label="Test Data", color='green')
plt.plot(test.index, forecast_mean, label="Forecasted Values", color='red', linestyle='dashed')
plt.fill_between(test.index, forecast_conf_int.iloc[:, 0], forecast_conf_int.iloc[:, 1], color='gray', alpha=0.2, label="Confidence Interval")
plt.title("Forecast vs Actuals")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend(loc="upper left")
plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Define SARIMA model with seasonal components (adjust p, d, q, P, D, Q, m)
sarima_model = SARIMAX(train,
                       order=(2, 0, 3),  # ARIMA order (p,d,q)
                       seasonal_order=(1, 1, 1, 12),  # Seasonal order (P,D,Q,m)
                       enforce_stationarity=False,
                       enforce_invertibility=False)

# Fit the model
sarima_fitted_model = sarima_model.fit()

# Forecast
sarima_forecast = sarima_fitted_model.get_forecast(steps=forecast_periods)
sarima_forecast_mean = sarima_forecast.predicted_mean

# Plotting the forecast
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label="Training Data", color='blue')
plt.plot(test.index, test, label="Test Data", color='green')
plt.plot(test.index, sarima_forecast_mean, label="Forecasted Values", color='red', linestyle='dashed')
plt.fill_between(test.index, sarima_forecast.conf_int().iloc[:, 0], sarima_forecast.conf_int().iloc[:, 1], color='gray', alpha=0.2, label="Confidence Interval")
plt.title("SARIMA Forecast vs Actuals")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend(loc="upper left")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(test, sarima_forecast_mean)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

print(f'MSE: {mse}')
print(f'RMSE: {rmse}')
```

```python
from sklearn.metrics import r2_score

# Calculate R^2 score for SARIMA
r2_sarima = r2_score(test, sarima_forecast_mean)
print("R² Score for SARIMA:", r2_sarima)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
```

```python
# Assume 'train' and 'test' are the dataframes you have for training and testing

# Scale the data (LSTM models perform better on normalized data)
scaler = MinMaxScaler(feature_range=(0, 1))  # Scale between 0 and 1
train_scaled = scaler.fit_transform(train.values.reshape(-1, 1))  # Reshaping for scaling
test_scaled = scaler.transform(test.values.reshape(-1, 1))

# Create dataset for LSTM input (sliding window approach)
def create_dataset(data, time_step=1):
    X, y = [], []
    for i in range(len(data) - time_step - 1):
        X.append(data[i:(i + time_step), 0])
        y.append(data[i + time_step, 0])
    return np.array(X), np.array(y)

# Set the time step (number of past observations used to predict the next one)
time_step = 30  # You can experiment with this value

X_train, y_train = create_dataset(train_scaled, time_step)
X_test, y_test = create_dataset(test_scaled, time_step)

# Reshape data to be in the form [samples, time steps, features] for LSTM
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

```

```python
# Initialize the LSTM model
model = Sequential()

# Add the first LSTM layer
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))

# Add dropout to prevent overfitting
model.add(Dropout(0.2))

# Add the second LSTM layer
model.add(LSTM(units=50, return_sequences=False))

# Add dropout again
model.add(Dropout(0.2))

# Add the output layer
model.add(Dense(units=1))  # Predicting a single value

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Fit the model
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
# Predict the future values on the test set
lstm_forecast_scaled = model.predict(X_test)

# Inverse transform the predicted values to get them back to the original scale
lstm_forecast = scaler.inverse_transform(lstm_forecast_scaled)
```

```python
# Ensure both x and y have the same length
plt.plot(test.index[time_step:time_step + len(lstm_forecast)], lstm_forecast, 
         label="LSTM Forecasted Values", color='red', linestyle='dashed')

# Adding title and labels
plt.title("LSTM Forecast vs Actuals")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend(loc="upper left")
plt.show()

```

```python
# Plot actual test data
plt.figure(figsize=(12, 6))
plt.plot(test.index, test, label="Actual Test Values", color='blue')

# Plot LSTM forecasted values
plt.plot(test.index[time_step:time_step + len(lstm_forecast)], lstm_forecast, 
         label="LSTM Forecasted Values", color='red', linestyle='dashed')

# Adding title and labels
plt.title("LSTM Forecast vs Actuals")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend(loc="upper left")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the 'temp' column to scale the data between 0 and 1
scaler = MinMaxScaler(feature_range=(0, 1))
df['scaled_temp'] = scaler.fit_transform(df[['temp']])

# Split the data into train and test sets
train_size = int(len(df) * 0.8)
train, test = df[:train_size], df[train_size:]

# Reshape the data for LSTM model (samples, timesteps, features)
X_train, y_train = [], []
X_test, y_test = [], []

def create_dataset(data, look_back=1):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back), 0])
        y.append(data[i + look_back, 0])
    return np.array(X), np.array(y)

look_back = 30  # Use the previous 30 days to predict the next day
train_data = train['scaled_temp'].values.reshape(-1, 1)
test_data = test['scaled_temp'].values.reshape(-1, 1)

X_train, y_train = create_dataset(train_data, look_back)
X_test, y_test = create_dataset(test_data, look_back)

# Reshaping input data for LSTM: (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Create the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=False, input_shape=(look_back, 1)))  # Use return_sequences=False if you want only the last output
model.add(Dense(units=1))

model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
# Make predictions on the test set
predictions = model.predict(X_test)

# Inverse transform the predictions and actual values to get them back to the original scale
predicted_temp = scaler.inverse_transform(predictions)
actual_temp = scaler.inverse_transform(y_test.reshape(-1, 1))

# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(test.index[look_back:], actual_temp, label='Actual Temperature', color='blue')
plt.plot(test.index[look_back:], predicted_temp, label='LSTM Forecasted Temperature', color='red', linestyle='dashed')
plt.title('LSTM Forecast vs Actual Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(actual_temp, predicted_temp)
print(f"Mean Squared Error (MSE): {mse}")

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python

# Calculate R^2 score for LSTM
r2_lstm = r2_score(actual_temp, predicted_temp)
print("R² Score for LSTM:", r2_lstm)
```
